```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
B = pd.read_csv('book (1).csv')
B
```

## min_support=0.1 'confidence'

```python
items = apriori(B ,min_support=0.1 ,use_colnames=True)
items
```

```python
rules = association_rules(items, metric='confidence', min_threshold=0.5)
rules
```

```python
rules.sort_values('confidence', ascending = False)
```

# PLOTS

```python
import matplotlib.pyplot as plt
plt.scatter(data=rules,x='support',y='confidence')
```

## 'lift'

```python
rules_lift = association_rules(items, metric='lift', min_threshold=0.5)
rules_lift
```

```python
rules.sort_values('lift', ascending=False)
```

```python
import matplotlib.pyplot as plt
plt.scatter(data = rules,x = 'support', y = 'lift')
```
